## 3.- Analysis numerical variable SalePrice

```python
df_train["SalePrice"].describe()
```

```python
sns.displot(df_train["SalePrice"])
```

```python
df_train["SalePrice"].skew() # If skew(dataFrame)=0, the distribution is symetry
                             # If skew(dataFrame)>0, the distribution is right skewed
                             # If skew(dataFrame)<0, the distribution is left skewed
```

```python
df_train["SalePrice"].kurt() #If kurt(df)>3, there are a large concentration of values around their mean.
                             #If kurt(df)=3, there are a normal concentration of values around their mean.
                             #If kurt(df)<3, there is a low concentration of values around their mean
```

## 3.- Analysis numerical variable vs categorial variable

#Now we are going to study categorical variable vs numerical variable

```python
data_3=pd.concat([df_train["SalePrice"],df_train["OverallQual"]],axis=1)
```

```python
data_3.head(10)
```

```python
f, ax =plt.subplots(figsize=(8,6))
fig=sns.boxplot(x="OverallQual",y="SalePrice",data=data_3)
fig.axis(ymin=0,ymax=800000)
```

```python
data_4=pd.concat([df_train["SalePrice"],df_train["YearBuilt"]],axis=1)
```

```python
data_4.head(5)
```

```python
f, ax =plt.subplots(figsize=(20,8))
fig=sns.boxplot(x="YearBuilt",y="SalePrice",data=data_4)
fig.axis(ymin=0,ymax=800000)
plt.xticks(rotation=90)
```
