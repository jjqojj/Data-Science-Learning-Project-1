## 3.- Analysis numerical variable SalePrice

```python
df_train["SalePrice"].describe()
```

```python
sns.distplot(df_train["SalePrice"])
```

```python
df_train["SalePrice"].skew()
```

```python
df_train["SalePrice"].kurt()
```

```python
# In the next, we want to build a new table with two variable from the origin table (original data frame)
```

```python
data=pd.concat([df_train["SalePrice"],df_train["GrLivArea"]], axis=1)
```

```python
data.head(10)
```

```python
data.plot.scatter(x="GrLivArea",y="SalePrice",ylim=(0,800000));
```

```python
data_2=pd.concat([df_train["SalePrice"],df_train["TotalBsmtSF"]], axis=1)
data_2.plot.scatter(x="TotalBsmtSF",y="SalePrice",ylim=(0,800000));
```

```python
#Now we are going to study categorical variable vs numerical variable
```

```python
data_3=pd.concat([df_train["SalePrice"],df_train["OverallQual"]],axis=1)
```

```python
data_3.head(10)
```

```python
f, ax =plt.subplots(figsize=(8,6))
fig=sns.boxplot(x="OverallQual",y="SalePrice",data=data_3)
fig.axis(ymin=0,ymax=800000)
```

```python
data_4=pd.concat([df_train["SalePrice"],df_train["YearBuilt"]],axis=1)
```

```python
data_4.head(5)
```

```python
f, ax =plt.subplots(figsize=(20,8))
fig=sns.boxplot(x="YearBuilt",y="SalePrice",data=data_4)
fig.axis(ymin=0,ymax=800000)
plt.xticks(rotation=90)
```
